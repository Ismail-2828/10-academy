```python

def preprocess(text):
    text = text.lower().strip()
    text = re.sub(r"[^a-z\s]","",text) #to rmv space and anything not lower
    text = [d for d in text.split() if d not in stopwords]
    lemmatizer = WordNetLemmatizer()
    text = [lemmatizer.lemmatize(text) for text in text]
    text = " ".join(text)
    return text

rating_df['title_processed'] = rating_df['title'].apply(preprocess)
rating_df['content_processed'] = rating_df['content'].apply(preprocess)
```
